```python
hay_nulos = df.isnull().values.any()
if hay_nulos:
    print("El DataFrame contiene valores nulos.")
else:
    print("El DataFrame no contiene valores nulos.")
```

```python
#al encontrar nulos limpiamos la base con dropna 
```

```python
df1= df.dropna()
hay_nulos_despues = df1.isnull().values.any()
if hay_nulos_despues:
    print("El DataFrame resultante aún contiene valores nulos después de eliminarlos.")
else:
    print("El DataFrame resultante no contiene valores nulos después de eliminarlos.")
```
